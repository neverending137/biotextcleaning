```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Assuming you've already downloaded the necessary nltk datasets

# Load your dataframe (assuming it's named con_data and the text column is named 'clean_bio')
# con_data = pd.read_csv("your_file.csv") # Uncomment and adjust as needed

# Initialize the stemmer
stemmer = PorterStemmer()

# Define a function to process the text
def remove_stopwords_and_stem(clean_bio):
    # Ensure that the input is a string
    if not isinstance(clean_bio, str):
        return clean_bio

    # Tokenize
    words = word_tokenize(clean_bio)

    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]

    # Stem the words
    stemmed_words = [stemmer.stem(word) for word in filtered_words]

    return ' '.join(stemmed_words)

# Apply the function to the text column and create a new column
con_data1['bio_text'] = con_data1['clean_text'].apply(remove_stopwords_and_stem)
con_data1['bio_text']
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(pd.DataFrame(con_data2_sorted['vec'].values.tolist()))
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in K:
    if k == 1:  # Silhouette score is not defined for a single cluster
        silhouette_scores.append(-1)
        continue
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(pd.DataFrame(con_data2_sorted['vec'].values.tolist()))
    score = silhouette_score(pd.DataFrame(con_data2_sorted['vec'].values.tolist()), labels)
    silhouette_scores.append(score)
```

```python
import matplotlib.pyplot as plt

# Plotting the sum of squared distances
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)
plt.plot(K, sum_of_squared_distances, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method')

# Plotting the silhouette scores
plt.subplot(1, 2, 2)
plt.plot(K, silhouette_scores, 'go-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.tight_layout()
plt.show()
```
